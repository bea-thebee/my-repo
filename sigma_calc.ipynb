```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as random
import scipy.stats as stats
import seaborn as sns

avg = []
stdev = []
err = []
```

```python
c = stats.chi2(1)
v = c.rvs(100).mean()
v
```

```python
from scipy.stats import norm
generator = norm()
N=100
exp = np.array([generator.rvs(N).std() for i in range(10000)])
plt.hist(exp, bins=50, density=True, histtype='stepfilled', alpha=0.5)
plt.plot((x:=np.linspace(-1,2,num=50)), norm.pdf(x, *norm.fit(exp)),
    label =rf'$\mu$={norm.fit(exp)[0]:.3f}'+'\n'+ rf'$\sigma$={norm.fit(exp)[1]:.3f}')
plt.legend(fontsize=10); plt.show()
```

```python
# assuming the intented means are both one, this is what is done
percent_error_chi2 = (np.abs(v - 1)/1) * 100
percent_error_likelihood = (np.abs(0.923 - 1)/1) * 100

percent_error_chi2, percent_error_likelihood
```
