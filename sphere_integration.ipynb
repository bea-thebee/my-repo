```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('Solarize_Light2')
import scipy.stats as stats

R, C = 100000, 3 #stands for rows and columns
rand = stats.uniform(-1, 2)
samp = rand.rvs(C*R).reshape(C, -1)
x, y, z = samp
```

```python
r = np.sqrt(x**2 + y**2 + z**2)

fig = plt.figure(figsize=(15,6))

#Had to change the implementation of the axes so that there could be a 3d projection of the scattered points

ax1 = plt.subplot(121, projection='3d')
ax1.scatter(x, y, z, c = "#371E5A")
ax1.set(xlabel='x', ylabel='y', zlabel='z')

ax2 = plt.subplot(122)
ax2.hist(r, bins=50, histtype='step', color='g')
ax2.set(xlabel='r')

ratio = sum(r<=1)/R #this is being used to calculate the volume of the sphere within the cube of points, or the actual integral
pi = sum(r<=1)/R*6 #and this is being used to calculate pi from that integral, using the pi/6 number that is present in volume of a sphere
print(f'For a sample size of {R:,} we can estimate the value as {ratio} and pi as {pi}')

plt.show()
```
