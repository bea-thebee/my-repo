```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random

sns.set(style="whitegrid")
plot = sns.displot(random.normal(size=1000), kde=False, bins=30)
plot;
```

```python
def log_likelihood(data):
    n = len(data)
    mu = np.mean(data)
    sigma = np.std(data, ddof=1)
    # the function below comes from the formula for log likelihood of a normal distribution. Sigma is stdev and mu is mean
    log_likelihood = -n/2 * np.log(2 * np.pi) - n * np.log(sigma) - np.sum((data - mu)**2) / (2 * sigma**2)
    return -log_likelihood

print(log_likelihood(plot.data))
```
