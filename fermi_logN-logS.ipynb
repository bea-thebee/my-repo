```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
plt.style.use('dark_background')

df = pd.read_csv('basic_4FGL_DR4.csv') #converting the fermi dataset into a dataframe
df = df.sort_values(by = 'association', ascending = False)
df.reset_index(inplace=True)
df = df.drop(columns='index') 
df = df.drop(np.arange(0,3206)) #after the df was sorted alphabetically, I removed everything from the df before blazars
df = df.sort_values(by = 'logS', ascending = False)
df.reset_index(inplace = True)
df = df.drop(columns = 'index')
df #the dataframe is now sorted by logS highest to lowest
```

```python
logN = np.log10(df.index)
df['logN'] = logN
df.iloc[1:] # removing top row because it's problematic
```

```python
x = df.logS
y = df.logN
known_line = -1.5*x -13.57
fig = plt.figure(figsize=[8,8])
ax1 = plt.subplot(111)
ax1.scatter(x,y, c = "#1B0BAEFF", s=10)
ax1.set(xlabel='logS', ylabel='log(N>S)')
ax1.plot(x, known_line, c = "#F9FAFA", linestyle= 'dotted')
ax1.set_ylim(ymax = 3.75)
```
