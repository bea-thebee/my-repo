```python
def weighted_mean(N, n, loc, low, high):
    array = np.array([norm.rvs(loc=loc, scale=(random.randint(low,high)), size=N) for i in range (n)])
    df = pd.DataFrame(array)
    stds = df.std(axis = 1)
    weights = pd.DataFrame(1 / stds**2)
    mean = np.sum((df * weights) / (weights * n))
    return mean
```

```python
def var(N, a1, a2, w1, w2):
    avg_sq_w = ((w1 + w2) / 2)**2
    avg_w_sq = (w1**2 + w2**2) / 2
    variance = (((np.sum(a1 - np.mean(a1)) 
      + np.sum(a2 - np.mean(a2))) 
      / (N-1)) 
      * (avg_w_sq / avg_sq_w))
    return variance

def mean_weighted(N, a1, a2, w1, w2):
    weighted_mean = (np.sum(a1 * w1) + np.sum(a2 * w2)) / ((w1 * N) + (w2 * N))
    return weighted_mean
```
