```python
 #The function I'm using to get pi and return other variables used. Prints are just to test things

import numpy as np
import scipy.stats as stats

def area_to_pi(square_area):
   circle_area = 0.7854 * square_area
   radius = np.sqrt(square_area) / 2
   my_pi = circle_area / (radius ** 2)
   return circle_area, radius, my_pi

print(area_to_pi(stats.uniform.rvs(loc=0, scale=100)))
```

```python
import numpy as np
import scipy.stats as stats
# this converts the volume of a cube to the volume of sphere using ratios, and then calculates pi from that volume
def volume_to_pi(cube_volume):
    sphere_volume = 0.5236* cube_volume
    radius = (cube_volume ** (1/3)) / 2
    my_pi = sphere_volume / (radius ** 3 * (4/3))
    return sphere_volume, radius, my_pi

print(volume_to_pi(stats.uniform.rvs(loc=0, scale=100)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
import numpy as np

all_stats = []
trials = 1000
pd.options.display.float_format = '{:,.16f}'.format

for i in range(trials):
    cube_volume = stats.uniform.rvs(loc=0, scale=100)
    sphere_volume, radius, my_pi = volume_to_pi(cube_volume)
    
    df = pd.DataFrame(index=range(trials), data={
        'Cube Volume' : [cube_volume],
        'Sphere Volume' : [sphere_volume],
        'Radius' : [radius],
        'My Pi' : [my_pi]
    })

    all_stats.append([
        df['Cube Volume'].sum(), 
        df['Sphere Volume'].sum(), 
        df['Radius'].sum(),
        df['My Pi'].sum()])
    
results_df = pd.DataFrame.from_records(all_stats, columns=['Cube Volume', 'Sphere Volume', 'Radius', 'My Pi'])
results_df.describe().style.format('{:,}')
```

```python
def percent_uncertainty(stdev, mean):
    return (stdev / mean) * 100

def percent_error(actual, my_value):
    return (abs(actual - my_value) / actual) * 100


actual = np.pi
my_value = results_df['My Pi'].mean() / 1000  # Adjusting for scale
stdev = results_df['My Pi'].std() / 1000
mean = results_df['My Pi'].mean() / 1000

print(f'Percent Uncertainty: {percent_uncertainty(stdev, mean)}', f'Percent Error: {percent_error(actual, my_value)}')
```
